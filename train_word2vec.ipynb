## Imports

```python
import pickle
import multiprocessing
from gensim.models import Word2Vec
import time
import numpy as np
```

```python
cores = multiprocessing.cpu_count()
print(cores)
```

## Load training data

```python
collection_path = "D:\\uni\\semester 7\\Information Retrieval\\Project\\IR_Code\\collection.obj"
```

```python
with open(collection_path, 'rb') as collection_file:
     collection = pickle.load(collection_file)
```

## Create word2vec model

```python
%%time
w2v_model = Word2Vec(min_count=1,
                     window=5,
                     vector_size=300,
                     alpha=0.03,
                     workers=cores-1)

w2v_model.build_vocab(collection)
w2v_model_vocab_size = len(w2v_model.wv)
print("vocab size: ", w2v_model_vocab_size)
```

```python
%%time
w2v_model.train(collection, total_examples=w2v_model.corpus_count, epochs=30)
```
